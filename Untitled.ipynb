## Missing data handling.

```python
# now for handling missing data thats in column age, cabin and emarked i will be using median for numerical columns and 
# mode for categorical columns.
# Define a function to fill missing values
# For numerical column first.
med= df["Age"].median()
# inplace ensures that all the changes are saved in the dataframe, fillna ensures that all the nan values are filled in the column or dataframe
df["Age"].fillna(med,inplace= True)
# Now checking if its all values are not missing and are really filled.
print("Missing values in 'age' column after imputation:")
print(df["Age"].isnull().sum())

# Fill missing values in the 'cabin' column with the mode
# i am using mode()[0]as a mode is a series and can have more than one modal values in it therefore by choosing 1 I am
# that all the missing values are being replaced by the first modal value10
mode_cabin = df['Cabin'].mode()[0]
df['Cabin'].fillna(mode_cabin, inplace=True)
# Check for missing values in the 'cabin' column after imputation
print("Missing values in 'cabin' column after imputation:")
print(df['Cabin'].isnull().sum())


# Fill missing values in the 'embarked' column with the mode
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)
# Check for missing values in the 'embarked' column after imputation
print("Missing values in 'embarked' column after imputation:")
print(df['Embarked'].isnull().sum())
```

## Data transformation

```python
# now first get .info() the data that has object in its type is categorical data
df.info()
# so after going through this .info i got to know that i have 5 categorical columns which are Name, sex, Ticket, Cabin and Embarked
# now i am going to trandform them to numerical data using one hot encoding.
df= pd.get_dummies(df, columns=['Name','Sex','Ticket','Cabin','Embarked'],dtype="float")
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the data
scaler.fit(df)

# Get the max values in the data
max_values = scaler.data_max_
print("Max values in the data:", max_values)

# Transform the data
normalized_data = scaler.transform(df)
print("Normalized data:")
print(normalized_data)

# Convert the normalized data back to a DataFrame
df_normalized = pd.DataFrame(normalized_data, columns=df.columns)
print(df_normalized)
```
